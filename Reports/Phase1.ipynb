Project Goals:
1. classification: we will train our model to classify individuals into different obesity levels such as Normal Weight, Overweight, and Obesity Type I, II, or III based on their eating habits and physical conditions, including attributes like age, weight, height, physical activity, and other related factors.
2. clustering: our model will create a set of clusters for individuals who share similar characteristics. These clusters will then be used to better understand patterns in the data and to provide insights that can help in predicting obesity levels and supporting decision-making for healthcare professionals.
