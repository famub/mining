# [1] Problem

**Obesity** is a major global health issue linked to serious conditions such as diabetes and cardiovascular disease. Early estimation of obesity levels is important because it helps identify at-risk individuals and supports preventive health decisions.
In this project, we aim to estimate obesity levels using data about **eating habits** and **physical conditions**. By analyzing these lifestyle behaviors, we seek to understand the **key factors** associated with obesity and develop **predictive models** that can support early intervention and promote healthier choices.
Therefore, the core task of this project is to apply data mining techniques to categorize individuals by their obesity level and group similar lifestyle profiles.

# [2] Data Mining Task

This project focuses on two main data mining tasks:
1. **Classification**
A **supervised learning** approach where the model will be trained to classify individuals into different **obesity level** categories such as Normal Weight, Overweight, and Obesity Type I, II, or III. The classification is based on a set of independent variables (features) related to eating habits and physical conditions, including attributes like age, weight, height, physical activity, and other related factors. The goal is to build a predictive model (Decision Tree) that can accurately assign an obesity level to a new instance for early risk assessment.
2. **Clustering**
An **unsupervised learning** approach where the K-means algorithm will be applied to create a set of clusters for individuals who share **similar characteristics** in their eating habits and physical conditions. These clusters will then be analyzed to discover hidden, natural groupings within the population. The insights gained from these patterns will help to better understand distinct lifestyle profiles and support decision-making for healthcare professionals in predicting obesity levels and targeting specific intervention strategies.

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy import stats

# Create a working copy of data
df_processed = df.copy()

# Clean column names (essential step)
df_processed.columns = df_processed.columns.str.strip().str.replace('"', '').str.replace(' ', '_')

print("--- Data Preprocessing Started ---")
print("Raw Dataset Shape:", df_processed.shape)
```

**Missing Values**

The dataset was examined for missing or null values using summary inspection and analytical checks.
No missing values were found in any attribute, so no imputation or row removal was required.

```python
import os

print("Original shape:", df.shape)

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

df_no_outliers = df.copy()

for col in numeric_cols:
    mean_val = df_no_outliers[col].mean()
    differences = abs(df_no_outliers[col] - mean_val)
    max_idx = differences.idxmax()
    df_no_outliers = df_no_outliers.drop(max_idx)
    print(f"Removed one outlier from column: {col}")

print("\nAfter removing outliers using Mean method on all numeric columns:")
print("New shape:", df_no_outliers.shape)

print("\nOriginal DataFrame:")
print(df.head(10))

print("\nDataFrame after removing outliers:")
print(df_no_outliers.head(10))

output_dir = 'Dataset'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
df_no_outliers.to_csv(os.path.join(output_dir, 'Preprocessed_dataset.csv'), index=False)
print("\n Saved as Preprocessed_dataset.csv")
```

**Outliers Removal**

Columns like Age, Height, and Weight are critical variables that can contain outliers which, if not removed, may affect the accuracy of models or analysis.

Using the mean method is a simple approach to identify outliers, where absolute differences between the values and the mean are calculated, and the value with the largest difference (outlier) is removed.

```python
from sklearn.preprocessing import LabelEncoder

binary_cols = ['family_history_with_overweight', 'FAVC', 'SMOKE', 'SCC']
mapping = {'yes': 1, 'no': 0}

for col in binary_cols:
    df_processed[col] = df_processed[col].replace(mapping).astype(int)

binary_true_false_cols = []
true_false_mapping = {'True': 1, 'False': 0}

for col in binary_true_false_cols:
    df_processed[col] = df_processed[col].replace(true_false_mapping).astype(int)

categorical_for_ohe = ['Gender', 'CAEC', 'CALC', 'MTRANS']
df_processed = pd.get_dummies(df_processed, columns=categorical_for_ohe, drop_first=True)

print(df_processed.head().T)

label_encoder = LabelEncoder()
df_processed['NObeyesdad_Encoded'] = label_encoder.fit_transform(df_processed['NObeyesdad'])
df_processed = df_processed.drop('NObeyesdad', axis=1)

print("\n--- Target Variable 'NObeyesdad' encoded using Label Encoding. ---")
```

```python
from sklearn.preprocessing import LabelEncoder

boolean_cols = df_processed.select_dtypes(include='bool').columns

for col in boolean_cols:
    df_processed[col] = df_processed[col].astype(int)

print("\n--- Boolean Columns (True/False) converted to 1/0. ---")
```

Many attributes in the dataset were categorical (e.g., Gender, family history, transportation, smoking).

```python
cols_to_scale = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

scaler = StandardScaler()
df_processed[cols_to_scale] = scaler.fit_transform(df_processed[cols_to_scale])

print("\nNumeric features scaled using Standardization (Z-Score).")
print(df_processed)
```

**Normalization**

Z-Score Standardization was applied to numerical columns like Age, Height, and Weight because they have different scales. For example, Age may range from 0 to 100, while Weight could range from 50 to 200. Z-Score Standardization ensures each feature has a mean of 0 and a standard deviation of 1, allowing the model to treat all features equally and preventing features with larger values from dominating the model's learning process.

**Classification Techniques**

We applied Decision Tree Classifiers using two different attribute selection measures:



*   Gini Index
    *   Implemented using DecisionTreeClassifier(criterion="gini") from sklearn.tree.
    *   Gini is used to measure node impurity and is commonly used due to its efficiency.


*   Entropy (Information Gain)

    *   Implemented using DecisionTreeClassifier(criterion="entropy").
    *   Entropy measures the reduction in uncertainty, making it suitable for datasets with multi-class targets.






Both models were trained and tested using three train-test partitions:
70â€“30, 80â€“20, and 90â€“10, following the project requirements.
The evaluation included: accuracy, confusion matrices, and comparison across partitions.


**Clustering Technique**

For each K, we computed:

*   Silhouette Score (from sklearn.metrics) to measure cluster separation
*   WCSS (Within-Cluster Sum of Squares) to evaluate compactness
*   Scatter plot visualization to visually inspect cluster separation

These metrics helped determine the best K based on both quantitative performance and visual clarity.

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import tree
import pickle
```

```python

# Load the preprocessed dataset
df = pd.read_csv('Preprocessed_dataset.csv')

print(f"Dataset Shape: {df.shape}")
print(f"\nFirst 5 rows:")
print(df.head())
```

```python
# Separate features (X) and target (y)
X = df.drop('NObeyesdad_Encoded', axis=1)
y = df['NObeyesdad_Encoded']

print(f"\nTarget distribution:")
print(y.value_counts().sort_index())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import tree
import pickle

# Define three different partition sizes
partition_configs = [
    {'test_size': 0.30, 'name': '70-30 Split'},
    {'test_size': 0.20, 'name': '80-20 Split'},
    {'test_size': 0.10, 'name': '90-10 Split'}
]

# Dictionary to store results
all_results = {}

print(" Partition ")
for config in partition_configs:
    train_pct = (1 - config['test_size']) * 100
    test_pct = config['test_size'] * 100
    print(f"  â€¢ {config['name']}: Train={train_pct:.0f}%, Test={test_pct:.0f}%")
```

Data Partitioning Strategy:
Three partition configurations are used to evaluate model stability and generalization:

70-30 Split: Provides the largest test set (30%) for robust performance evaluation with high statistical confidence.

80-20 Split: The standard balanced approach, offering an optimal compromise between training data size and test reliability, recommended for general use.

90-10 Split: Maximizes training data (90%) to achieve highest potential accuracy and optimal feature learning, though the smaller test set (10%) provides lower statistical confidence.

This multi-configuration approach ensures comprehensive model assessment across varying data distributions.

```python

# Train models with different partitions and criteria
for config in partition_configs:
    test_size = config['test_size']
    split_name = config['name']

    print(f"\n{'â”€' * 80}")
    print(f" {split_name}")
    print(f"{'â”€' * 80}")

    # Split the data with fixed random_state for reproducibility
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42, stratify=y
    )

    print(f"Training samples: {len(X_train)}")
    print(f"Test samples: {len(X_test)}")

    # Train with GINI criterion

    clf_gini = DecisionTreeClassifier(random_state=42)
    clf_gini.fit(X_train, y_train)


    # Train with ENTROPY criterion (Information Gain)
    clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)
    clf_entropy.fit(X_train, y_train)


    # Store models and data
    all_results[split_name] = {
        'X_train': X_train,
        'X_test': X_test,
        'y_train': y_train,
        'y_test': y_test,
        'gini_model': clf_gini,
        'entropy_model': clf_entropy
    }
```


The analysis aims to compare the performance of Decision Tree models using different data splits (70-30, 80-20, 90-10) and criteria (Gini vs. Entropy) to determine which configuration provides the most accurate and reliable classification, forming the basis for final recommendations.

```python

summary_data = []

for split_name, data in all_results.items():


    X_test = data['X_test']
    y_test = data['y_test']

    # Evaluate GINI model

    y_pred_gini = data['gini_model'].predict(X_test)
    accuracy_gini = accuracy_score(y_test, y_pred_gini)


    # Evaluate ENTROPY model

    y_pred_entropy = data['entropy_model'].predict(X_test)
    accuracy_entropy = accuracy_score(y_test, y_pred_entropy)


    # Store predictions for later use
    data['y_pred_gini'] = y_pred_gini
    data['y_pred_entropy'] = y_pred_entropy

    summary_data.append({
        'Split': split_name,
        'Criterion': 'Gini',
        'Accuracy': f"{accuracy_gini:.4f}",
        'Test Samples': len(y_test)
    })
    summary_data.append({
        'Split': split_name,
        'Criterion': 'Entropy',
        'Accuracy': f"{accuracy_entropy:.4f}",
        'Test Samples': len(y_test)
    })


summary_df = pd.DataFrame(summary_data)
print("\n", summary_df.to_string(index=False))
```


Entropy consistently outperforms Gini across all splits, indicating it is more effective at maximizing Information Gain and node purity. The highest accuracy is achieved with the Entropy 90-10 split (0.9717), and overall performance improves as the training set size increases, showing that the model benefits from more

```python
import os

os.makedirs('Dataset', exist_ok=True)

# Create confusion matrices plot
fig, axes = plt.subplots(3, 2, figsize=(12, 16))
fig.suptitle('Confusion Matrices for All Models', fontsize=16, fontweight='bold')

plot_idx = 0
for split_name, data in all_results.items():
    y_test = data['y_test']

    # Confusion matrix for GINI
    cm_gini = confusion_matrix(y_test, data['y_pred_gini'])
    ax = axes[plot_idx // 2, 0]
    sns.heatmap(cm_gini, annot=True, fmt='d', cmap='Blues', ax=ax,
                cbar=True, square=True)
    ax.set_title(f'{split_name} - Gini Index', fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

    # Confusion matrix for ENTROPY
    cm_entropy = confusion_matrix(y_test, data['y_pred_entropy'])
    ax = axes[plot_idx // 2, 1]
    sns.heatmap(cm_entropy, annot=True, fmt='d', cmap='Greens', ax=ax,
                cbar=True, square=True)
    ax.set_title(f'{split_name} - Entropy (IG)', fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

    plot_idx += 2

plt.tight_layout()
plt.savefig('Dataset/confusion_matrices.png', dpi=300, bbox_inches='tight')
plt.show()

# Print confusion matrices as text
for split_name, data in all_results.items():
    print(f"\n{split_name}:")
    print(f"\nGini Index Confusion Matrix:")
    print(confusion_matrix(data['y_test'], data['y_pred_gini']))
    print(f"\nEntropy Confusion Matrix:")
    print(confusion_matrix(data['y_test'], data['y_pred_entropy']))
```

70-30 Split:
Entropy outperforms Gini Index with 94.16% accuracy versus 90.85% (3.31% improvement). The confusion matrix shows Entropy achieved fewer misclassifications, particularly in Class 1 where Gini had 12 errors to Class 5 while Entropy had only 6.

80-20 Split:
Entropy maintains superior performance at 94.33% compared to Gini's 91.02% (3.31% difference). Error reduction is consistent across class pairsâ€”Class 0 to Class 1 misclassifications decreased from 7 to 3, and Class 1 to Class 5 from 6 to 2.

90-10 Split:
Entropy achieves peak performance at 97.17% versus Gini's 91.51% (5.66% improvement). With 90% training data, Entropy produced only 5 misclassifications compared to Gini's 18, demonstrating superior learning capacity.

```python
import numpy as np

splits = list(all_results.keys())

all_metrics = {}

print("\n" + "=" * 80)
print(" " * 25 + "DECISION TREE ANALYSIS")
print("=" * 80)

for split_name in splits:
    print(f"\n\n{split_name.upper()}")
    print("-" * 80)

    split_data = all_results[split_name]

    y_test = split_data['y_test']
    y_pred_gini = split_data['y_pred_gini']
    y_pred_entropy = split_data['y_pred_entropy']

    correct_gini = sum(y_test == y_pred_gini)
    total = len(y_test)
    accuracy_gini = correct_gini / total

    correct_entropy = sum(y_test == y_pred_entropy)
    accuracy_entropy = correct_entropy / total

    print(f"\nGINI INDEX:")
    print(f"  Accuracy:        {accuracy_gini:.4f} ({accuracy_gini*100:.2f}%)")
    print(f"  Correct/Total:   {correct_gini}/{total}")

    print(f"\nENTROPY (Information Gain):")
    print(f"  Accuracy:        {accuracy_entropy:.4f} ({accuracy_entropy*100:.2f}%)")
    print(f"  Correct/Total:   {correct_entropy}/{total}")

    diff = abs(accuracy_gini - accuracy_entropy)
    better = "GINI" if accuracy_gini > accuracy_entropy else "ENTROPY" if accuracy_entropy > accuracy_gini else "EQUAL"

    print(f"\nComparison:")
    print(f"  Difference:      {diff:.4f} ({diff*100:.2f}%)")
    print(f"  Better Method:   {better}")

    all_metrics[split_name] = {
        'gini': {'accuracy': accuracy_gini},
        'entropy': {'accuracy': accuracy_entropy}
    }

print(f"\n\n{'=' * 80}")
print(" " * 25 + "OVERALL COMPARISON")
print("=" * 80)

print(f"\nGINI INDEX RESULTS:")
print("-" * 80)
print(f"{'Split':<20} {'Accuracy':>15} {'Correct/Total':>20} {'Percentage':>15}")
print("-" * 80)

for split_name in splits:
    metrics = all_metrics[split_name]['gini']
    split_data = all_results[split_name]
    total = len(split_data['y_test'])
    correct = sum(split_data['y_test'] == split_data['y_pred_gini'])

    print(f"{split_name:<20} {metrics['accuracy']:>15.4f} {f'{correct}/{total}':>20} {metrics['accuracy']*100:>14.2f}%")

print(f"\n\nENTROPY (INFORMATION GAIN) RESULTS:")
print("-" * 80)
print(f"{'Split':<20} {'Accuracy':>15} {'Correct/Total':>20} {'Percentage':>15}")
print("-" * 80)

for split_name in splits:
    metrics = all_metrics[split_name]['entropy']
    split_data = all_results[split_name]
    total = len(split_data['y_test'])
    correct = sum(split_data['y_test'] == split_data['y_pred_entropy'])

    print(f"{split_name:<20} {metrics['accuracy']:>15.4f} {f'{correct}/{total}':>20} {metrics['accuracy']*100:>14.2f}%")

```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# load pre-processed data
df_clus = pd.read_csv("Preprocessed_dataset.csv")

# drop the class label
features = df_clus.drop('NObeyesdad_Encoded', axis=1)

# scale all features
scaler = StandardScaler()
scaled_features= scaler.fit_transform(features)

# new DataFrame with the scaled features
df_scaled = pd.DataFrame(scaled_features, columns=features.columns)

# Display the scaled DataFrame head
print("\nScaled DataFrame Head with Pre-processed Data:")
print(df_scaled.head())
```

```python
from sklearn.cluster import KMeans

wcss = []

for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_features)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters (K)")
plt.ylabel("WCSS")
plt.show()
```

The WCSS decreases as K increases, and the curve begins to flatten around K=2â€“4.
 This suggests that K=2 is a reasonable choice since adding more clusters gives only small improvement.

```python
k_values = [2, 3, 4]
silhouette_scores = []
wcss_k = []
```

```python
from sklearn.cluster import KMeans

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    labels = kmeans.labels_

    print("Cluster Centers")
    print(kmeans.cluster_centers_)
    print("\nCluster Labels:")
    print(labels)
```

Evaluation was done using Silhouette Score and WCSS.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

print("Silhouette Scores:\n")
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_features)

    sil = silhouette_score(scaled_features, labels)
    silhouette_scores.append(sil)

    print(f"K = {k} --> Silhouette = {sil}")

print("\nWCSS:\n")
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    wcss_k.append(kmeans.inertia_)

    print(f"K = {k} --> WCSS = {kmeans.inertia_}")
```

*   K = 2 gives the highest Silhouette Score, indicating the best separation between clusters.
*   K = 4 reduces WCSS, as expected when increasing the number of clusters.
*   K = 3 performs poorly with the lowest silhouette score

Based on the majority rule, the best K is the one that performs well across most metrics

**Optimal number of clusters: K = 2**

```python
plt.plot(k_values, silhouette_scores, marker='o')
plt.title("Silhouette Score for different K")
plt.xlabel("Number of clusters (K)")
plt.ylabel("Silhouette Score")
plt.show()
```

K=2 gives the highest Silhouette Score (â‰ˆ0.226), while K=3 and K=4 have lower values.
Therefore, K=2 provides the best cluster cohesion and separation.
